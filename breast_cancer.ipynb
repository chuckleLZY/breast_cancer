# Processing Data
Data Structure is **patient_id, path, target**

```python
base_path="/home/jupyter/kaggle/input/breast-histopathology-images/IDC_regular_ps50_idx5/"
patient = listdir(base_path)
len(patient)
```

```python
total_images = 0
for n in range(len(patient)):
    patient_id = patient[n]
    for c in [0, 1]:
        patient_path = base_path + patient_id 
        class_path = patient_path + "/" + str(c) + "/"
        subfiles = listdir(class_path)
        total_images += len(subfiles)
total_images
```

```python
data = pd.DataFrame(index=np.arange(0, total_images), columns=["patient_id", "path", "target"])

k = 0
for n in range(len(patient)):
    patient_id = patient[n]
    patient_path = base_path + patient_id 
    for c in [0,1]:
        class_path = patient_path + "/" + str(c) + "/"
        subfiles = listdir(class_path)
        for m in range(len(subfiles)):
            image_path = subfiles[m]
            data.iloc[k]["path"] = class_path + image_path
            data.iloc[k]["target"] = c
            data.iloc[k]["patient_id"] = patient_id
            k += 1  

data.target = data.target.astype(np.int64)
data.head()
```

#### Customize the Dataset 

```python
class BreastCancerDataset(Dataset):
    
    def __init__(self, df, transform=None):
        """
        Args:
            states (dataframe): whole dataset information
            transform (callable, optional): transformation of the image
        """
        self.states = df
        self.transform=transform
      
    def __len__(self):
        return len(self.states)
        
    def __getitem__(self, idx):
        image_path = self.states.path.values[idx] 
        image = Image.open(image_path)
        image = image.convert('RGB')
        
        if self.transform:
            image = self.transform(image)
        
        target = np.int64(self.states.target.values[idx])
            
        return {'image': image,
                'label': target}
```

```python
# Show the first 50 pictures.
def display_pictures(dataset, idx):
    if len(idx)>50:
        size=50
    else:
        size=len(idx)
    fig, ax = plt.subplots(5,10,figsize=(20,10))
    for m in range(size):
        image = imread(dataset.states.path.values[m])
        ax[m//10,m%10].imshow(image)
        ax[m//10,m%10].grid(False)
        
```

```python
basic_transform = transforms.Compose([
    transforms.Resize((50, 50)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = BreastCancerDataset(train_df,transform=basic_transform)
# test_train_dataset = BreastCancerDataset(test_train_df, transform=basic_transform)
# test_dev_dataset = BreastCancerDataset(test_dev_df, transform=basic_transform)
```
